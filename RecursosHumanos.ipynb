### Transformando a coluna Attrition, onde yes= 1 e no= 0.

```python
bd["Attrition"] = bd['Attrition'].apply(lambda x: 1 if x =='Yes' else 0)
```

```python
bd["OverTime"] = bd['OverTime'].apply(lambda x: 1 if x =='Yes' else 0)
```

### Removendo colunas que não serão utilizadas.

```python
bd.drop(["EmployeeCount", "StandardHours", "Over18", "EmployeeNumber"], axis= 1, inplace= True)
bd.head()
```

```python
sairam = bd[bd["Attrition"] == 1]
permaneceram = bd[bd["Attrition"] == 0]
```

```python
print("Total= ", len(bd))
print("Número de funcionários que saíram da empresa= ", len(sairam))
print("Porcentagem de funcionários que saíram da empresa= ", (len(sairam)/len(bd))*100)
print("Número de funcionários que permaneceram na empresa= ", len(permaneceram))
print("Porcentagem de funcionários que permaneceram na empresa= ", (len(permaneceram)/len(bd))*100)      
```
